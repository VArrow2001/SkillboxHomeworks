# 6.1 Применение функций и метод apply

```python
import pandas as pd
```

```python
import numpy as np
```

```python
df = pd.read_csv("WeightLoss.csv")
```

```python
df
```

```python
df['total'] = df.w1 + df.w2 + df.w3
```

```python
df['total_gr'] = df.total * 1000
```

```python
df
```

```python
df['total_gr'].apply(np.log)
```

```python
df.head()
```

```python
df['av_loss'] = df.loc[:, 'w1':'w3'].apply(np.mean, axis = 1) #axis 1 делает по строке, 0 - по столбцу
```

```python
df.head()
```

```python
f = lambda x: x.max() - x.min()
```

```python
df['wrange'] = df.loc[:, 'w1' : 'w3'].apply(f, axis = 1)
```

```python
df.head()
```

# 6.2 Группировка и агрегирование

```python
list(df.groupby('group'))[0]
```

```python
df.groupby('group').agg('sum') #аггрегация, выводит функции по группам
```

```python
df.groupby('group').agg(['min', 'max', 'mean']).loc[:, 'w1']
```

# 6.3 Сортировка и упорядочение

```python
df.sort_values('total') #сортирока по группе total
```

```python
df.sort_values(['se3', 'total'], ascending = False) #приоритетность сортировки; ascending = False - по убыванию
```
