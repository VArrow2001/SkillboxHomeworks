```python
# Переименование столбца с целью дальнейшего объединения с другим датасетом
progresses.rename(columns = {'id':'progress_id'}, inplace = True)
```

```python
# Формирование словаря id_title для удобства
id_title = dict()
for i in range(15):
    id_title[courses.iloc[i].course_id] = courses.iloc[i].title
```

### Подсчет количества курсов

```python
courses_len = len(courses)
print(f"Всего курсов: {courses_len}")
```

```python
#Объединение датасета курсов с датасетом их модулей
courses = pd.merge(courses, course_contents)
```

### Подсчёт количества модулей в каждом курсе

```python
id_modnum = dict()
for i in id_title:
    id_modnum[i] = len(set(courses[courses.course_id == i].module_title))
    print(f"Количество модулей в курсе '{id_title[i]}':", id_modnum[i])
```

### Подсчёт количества уроков в каждом курсе

```python
for i in id_title:
    print(f"Количество уроков в курсе {id_title[i]} : \
{len(courses[courses['course_id'] == i])}")
```

### Подсчёт количества уроков в каждом модуле каждого курса

```python
course_modles = dict()
for course_id in id_title:
    print(f"\n\n==--Курс {id_title[course_id]}.--==\n\n Количество уроков\
в следующих модулях:")
    course_modles[course_id] = []
    for module_number in range(1, id_modnum[course_id] + 1):
        module_title = list(courses[(courses['module_number'] == 
        module_number) & (courses['course_id'] == 
        course_id)].module_title)[0]
        
        lessons_number = len(courses[(courses['module_number'] == 
        module_number) & (courses['course_id'] == 
        course_id)])
        
        course_modles[course_id].append(lessons_number)
        print(f"\t'{module_title}': \
{lessons_number}")
        
```

### Медианное количество уроков в модуле за каждый курс

```python
import numpy as np
for course_id in course_modles:
    print(f"{id_title[course_id]} : {np.median(course_modles[course_id])}")
```

### Подсчёт характеристик возраста всех студентов

```python
def age(birthday):
    if (str(birthday) != 'nan'):
        if 12<= (2019 - int(birthday[:4])) <= 100:
            return 2019 - int(birthday[:4])
    return None
```

```python
students['ages'] = students.birthday.apply(age)
```

```python
correct_ages = list(map(int, students.ages.dropna()))
```

```python
# Максимальное значение
max_age = max(correct_ages)
# Минимальное значение
min_age = min(correct_ages)
# Среднее значение
mean_age = round(np.mean(correct_ages), 2)
# Медианное значение
median_age = np.median(correct_ages)

print(f"Максмальный возраст: {max_age};\n\
Минимальный возраст: {min_age};\n\
Средний возраст: {mean_age};\n\
Медианный возраст: {median_age}.")
```

### Подсчёт характеристик возраста студентов каждого курса

```python
main = pd.merge(progresses, progress_phases)
```

```python
main = pd.merge(courses, main)
```

```python
del main['Unnamed: 0']
```

```python
course_studnum = dict()
for course in id_title.values():
    print(f'===Возрастные характеристики студентов курса "{course}"===\n\n')
    studs = list(set(main[main.title == course ]['student_id']))
    course_studnum[course] = len(studs)
    ages_list = [float(students[students.id == i].ages) for i in studs]
    ages_list = [i for i in ages_list if str(i) != 'nan']
    # Максимальное значение
    max_age = max(ages_list)
    # Минимальное значение
    min_age = min(ages_list)
    # Среднее значение
    mean_age = round(np.mean(ages_list), 2)
    # Медианное значение
    median_age = np.median(ages_list)
    print(f"Максмальный возраст: {max_age};\n\
Минимальный возраст: {min_age};\n\
Средний возраст: {mean_age};\n\
Медианный возраст: {median_age}.\n\n\n")
```
