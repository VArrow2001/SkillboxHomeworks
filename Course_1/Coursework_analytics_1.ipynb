### 1.0 Первичная обработка данных

```python
import pandas as pd

# Чтение каждого сsv файла и их запись в dataframe объекты
courses = pd.read_csv("Data\courses.csv")
course_contents = pd.read_csv("Data\course_contents.csv")
progress_phases = pd.read_csv("Data\progress_phases.csv")
progresses = pd.read_csv("Data\progresses.csv")
students = pd.read_csv("Data\students.csv")

# Переименование столбцов с целью дальнейшего объединения с другими датасетами

progresses.rename(columns = {'id':'progress_id'}, inplace = True)
courses.rename(columns = {'id':'course_id'}, inplace = True)
N = len(courses)
```

```python
#Создание словаря {id_курса:название_курса}
id_title = {courses.iloc[i].course_id:courses.iloc[i].title
            for i in range(N)}
id_title
```

```python
#Создание словаря {название_курса:сфера_курса}
course_field = {courses.iloc[i].title : courses.iloc[i].field
                for i in range(N)}
course_field
```

```python
#Объединение датасета курсов с датасетом их модулей
courses = pd.merge(courses, course_contents)
del courses['Unnamed: 0']
courses.head()
```

```python
#Объединеие всех датасеов, кроме students
main = pd.merge(progresses, progress_phases)
main = pd.merge(courses, main)
main.head()
```

**Примечание:**<br>
*Отдельный датасет для возраста студентов не создавал, так как получилось в той же таблице выполнить рассчеты по возрасту (см. далее).*

### 1.1 Рассчёт показателей

### 1.1.1 Количество курсов в датасете

### 1.1.2 Количество модулей на каждом курсе

```python
directory = 'CreatedData/'
```

```python
filename = 'NModulesPerCourse.html'
title_modnum = dict() # Словарь {название_курса:количество_модулей}
mod_p_course = pd.DataFrame(columns = ('Название курса', 'Количество модулей'))
for i in id_title.values():
    title_modnum[i] = len(set(courses[courses.title == i].module_title))
    mod_p_course.loc[len(mod_p_course)] = (i, title_modnum[i])
mod_p_course.to_html(directory + filename, index = False)
```

### 1.1.3 Количество уроков в каждом модуле на каждом курсе


```python
filename = 'NLessonsPerModule.html'
les_p_mod = pd.DataFrame(columns = ('Название курса', 'Название модуля',
                                   'Количество уроков'))
course_modules = dict() # Словарь {название_курса:[кол-во_уроков_помодульно]}
for course in id_title.values():
    course_modules[course] = []
    for module_number in range(1, title_modnum[course] + 1):
        module_title = list(courses[(courses['module_number'] == 
        module_number) & (courses['title'] == 
        course)].module_title)[0]
        
        lessons_number = len(courses[(courses['module_number'] == 
        module_number) & (courses['title'] == 
        course)])
        
        course_modules[course].append(lessons_number)
        les_p_mod.loc[len(les_p_mod)] = (course, module_title, lessons_number)
        #print(f"\t'{module_title}': \
#{lessons_number}")
les_p_mod.set_index(['Название курса', 'Название модуля'], inplace = True)
les_p_mod.to_html(directory + filename)
```

### 1.1.4 Медианное количество уроков в модуле на каждом курсе

```python
from numpy import median
filename = 'MedLessonsPerModule.html'
course_p_med = pd.DataFrame(columns = ('Название курса', 
                                       'Медианное кол-во уроков в модуле'))
for i in course_modules:
    course_p_med.loc[len(course_p_med)] = (i, median(course_modules[i]))
course_p_med.to_html(directory+filename, index = False)
```

### 1.1.5 Подсчёт характеристик возрастов всех студентов

```python
def age(birthday): # функция для подсчёта возраста студентов из даты рождения
    YEAR = 2020
    if (str(birthday) != 'nan'):
        if 12<= (YEAR - int(birthday[:4])) <= 100:
            return YEAR - int(birthday[:4])
    return None
```

```python
students['ages'] = students.birthday.apply(age) # создание столбца ages
students.dropna().head()
```

```python
# Отбор валидных значений возраста
correct_ages = list(map(int, students.ages.dropna()))
```

```python
from numpy import mean
filename = 'InfoStudAll.html'
char_value = pd.DataFrame(columns = ('Характеристика', 'Значение'))
char_value.loc[0] = ('Максимальный возраст', max(correct_ages))
char_value.loc[1] = ('Минимальный возраст', min(correct_ages))
char_value.loc[2] = ('Средний возраст', round(mean(correct_ages), 2))
char_value.loc[3] = ('Медианный возраст', median(correct_ages))
char_value.to_html(directory + filename, index = False)
```

### 1.1.6 Подсчёт характеристик возрастов студентов каждого курса

```python
course_studnum = dict() # словарь {название_курса:кол-во_студентов}
filename = 'InfoStudPerCourse.html'
course_stud_info = pd.DataFrame(columns = ('Название курса', 
                                           'Количество студентов',
                                           'Максимальный возраст',
                                           'Минимальный возраст',
                                           'Средний возраст',
                                           'Медианный возраст'))
for course in id_title.values():
    studs = list(set(main[main.title == course ]['student_id']))
    course_studnum[course] = len(studs)
    ages_list = [float(students[students.id == i].ages) for i in studs]
    ages_list = [i for i in ages_list if str(i) != 'nan']
    # Максимальное значение
    max_age = max(ages_list)
    # Минимальное значение
    min_age = min(ages_list)
    # Среднее значение
    mean_age = round(mean(ages_list), 2)
    # Медианное значение
    median_age = median(ages_list)
    course_stud_info.loc[len(course_stud_info)] = (course, len(studs), max_age, min_age, mean_age, median_age)
course_stud_info.to_html(directory + filename, index = False)
```

### 1.2 Горизонтальный bar-chart Student = f(Course)

```python
import matplotlib.pyplot as plt

%matplotlib inline
fig, subplot = plt.subplots()
x_ticks, y_ticks = list(course_studnum.keys()), list(course_studnum.values())
subplot.barh(x_ticks, y_ticks)
plt.savefig('CreatedData/course_stud.png', bbox_inches = 'tight')
```

### 1.3 Горизонтальный bar-chart Student = f(Course) с оформлением

```python
# Создаём словарь {сфера:цвет} для дальнейшего окрашивания каждого bar-а
field_color = {'Development': '#1589FF', 
               'Design': '#98FF98', 
               'Business': '#FFA62F', 
               'Marketing': '#F9A7B0'}
```

```python
# Создаём словарь, где значения отсортированы по возрастанию
sorted_dict = {j:i for i, j in course_studnum.items()}
sorted_dict = {sorted_dict[i]:i for i in sorted(sorted_dict)}
```

```python
# Создаём списки обозначений осей
x_ticks, y_ticks = list(zip(*list(sorted_dict.items())))
```

```python
# Создаём список цветов, который соответствует словарю field_color и
# отсортированным значениям x_ticks
color_array = [field_color[course_field[i]] for i in x_ticks]
```

```python
# Определяем медиану для дальнейшего построения её линии
med = median(y_ticks)
```

```python
from matplotlib.patches import Rectangle # используется для обозначений легенды
fig, subplot = plt.subplots()

# В этом фрагменте стирается рамка
subplot.spines['top'].set_visible(False)
subplot.spines['right'].set_visible(False)
subplot.spines['bottom'].set_visible(False)
subplot.spines['left'].set_visible(False)

# Строим график
subplot.barh(x_ticks, y_ticks, color = color_array, alpha = 0.1)

# Устанавливаем заголовок
subplot.set_title('Количество студентов на каждом курсе')

# Создаём кортеж окрашенных прямоугольников для использования в легенде
rect = tuple(Rectangle((0,0), 10., 5., 
                       color = field_color[i]) for i in field_color)

# Создаём легенду
subplot.legend(rect, tuple(field_color.keys()))

# Рисуем линию медианы
subplot.vlines(med,-0.4,14.4, color = 'brown')

plt.savefig('CreatedData/course_stud_format.png', bbox_inches = 'tight')
```

### 2.1 Рассчёт прироста студентов на каждом курсе

```python
directory = 'CreatedData2/'
```

```python
#Создадим список нужных месяцев
months = [str(i) + '-' + (j < 10) * '0' + str(j) for i in range(2016, 2020) for j in range(1, 13)]
months = months[months.index('2016-03'):months.index('2019-09')]
month_growth = pd.DataFrame(columns = ('Месяцы', *list(id_title.values())))
month_growth['Месяцы'] = months[:-1]
month_growth.head()
```

```python
filename = 'MonthlyGrowth.html'
for course in id_title.values():
    growth = dict() # словарь {'год-месяц':прирост}
    are_homeworks = courses[(courses.title == course) & (courses.is_homework == True)]
    n = len(are_homeworks)
    start_end = ((are_homeworks.iloc[0].module_number,are_homeworks.iloc[0].lesson_number),
                 (are_homeworks.iloc[n - 1].module_number, are_homeworks.iloc[n - 1].lesson_number))
    course_df = main[main.title == course]
    for i, el in enumerate(months[:-1]):
        cur_month = el
        next_month = months[i + 1]
        month_diap = lambda x: (x > cur_month) & (x < next_month)
        # отбор количества начавших и закончивших в этом месяце 
        growth[el] = len(course_df[(course_df.module_number == start_end[0][0]) & (course_df.lesson_number == start_end[0][1])
                           & (month_diap(course_df.start_date))])
    
    month_growth[course] = growth.values()
month_growth.to_html(directory + filename, index = False)
month_growth.tail()
```

### 2.2 График ежемесячного прироста студентов на каждом курсе

```python
from pandas.plotting import register_matplotlib_converters as rmc

rmc()
i = 0
for course in id_title.values():
    i += 1
    filename = 'Course_growth_' + str(i) + '.png'
    fig, subplot = plt.subplots()
    fig.set_size_inches(10, 5)
    subplot.plot(month_growth['Месяцы'], month_growth[course])
    subplot.set_title(f"Ежемесячный прирост студентов курса '{course}'")
    subplot.set_ylabel('Прирост')
    subplot.set_xlabel('Месяцы')
    plt.xticks(month_growth['Месяцы'], rotation = 'vertical')
    plt.savefig(directory + filename, bbox_inches = 'tight')
```

```python
filename = 'MonthlyProgress.html'
month_progress = pd.DataFrame(columns = ('Месяцы', *list(id_title.values())))
month_progress['Месяцы'] = months[:-1]
month_progress.head()

# функция для упрощения даты до формата "год-месяц"
cut_month = lambda x: str(x)[:7] 

# создание столбцов с упрощенными датами
main['start_month'] = main['start_date'].apply(cut_month)
main['finish_month'] = main['finish_date'].apply(cut_month)

# отбор значений, соответствующих выполненным домашним работам
cur_frame = main[(main.status == 'done') & (main.is_homework == True)]
for month in range(len(months) - 1):
    month_frame = cur_frame[(months[month] >= main.start_month) & (months[month] <= main.finish_month)]
    for course in id_title.values():
        month_progress[course][month] = len(month_frame[month_frame.title == course])
month_progress.to_html(directory + filename, index = False)
month_progress.tail()
```
